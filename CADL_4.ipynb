```python

# Import libraries
import nltk
import gensim
import gensim.corpora as corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import spacy
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Download resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Example corpus
documents = [
    "Machine learning is transforming the way we work and live.",
    "Artificial intelligence is a branch of computer science.",
    "Deep learning algorithms require large datasets.",
    "Natural language processing is a key component of AI systems.",
    "Data science and machine learning help businesses gain insights."
]

# Initialize stop words
stop_words = set(stopwords.words('english'))

# Preprocessing function
def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return tokens

# Apply preprocessing
texts = [preprocess(doc) for doc in documents]

# Create Dictionary and Corpus
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Apply LDA
lda_model = LdaModel(corpus=corpus,
                     id2word=dictionary,
                     num_topics=3,
                     random_state=100,
                     update_every=1,
                     passes=10,
                     alpha='auto',
                     per_word_topics=True)

# Print topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")

# Visualize the topics
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
